```python
from keybert import KeyBERT
import nltk
from nltk.stem import PorterStemmer,SnowballStemmer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from google.colab import drive
#drive.mount('/content/drive')

def keyword_extraction(text, top_n):
    kw_model = KeyBERT()
    keywords = kw_model.extract_keywords(str(text), keyphrase_ngram_range=(1, 1), stop_words='english', top_n=top_n*2)
    stemmer = SnowballStemmer("english")
    unique_keywords = []
    seen_stems = set()
    for keyword, score in keywords:
        stem = stemmer.stem(keyword)
        if stem not in seen_stems and len(unique_keywords) < top_n:
            unique_keywords.append(keyword)
            seen_stems.add(stem)
    keyword_scores = {keyword: score for keyword, score in keywords}

    return unique_keywords, keyword_scores

with open('/content/sample_data/text.txt', 'r', encoding='utf-8') as file:
    text = file.read()

top_n = int(input("Enter the number of keywords you want to be displayed: "))
keywords, keyword_scores = keyword_extraction(text, top_n)
print(keywords)
word_frequencies = {keyword: keywords.count(keyword) for keyword in keywords}
wordcloud = WordCloud(width=800, height=800, background_color='white').generate_from_frequencies(word_frequencies)
plt.figure(figsize=(10, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
